```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Loading the dataframe
flights = pd.read_csv('flights.csv', header=0, engine='python', encoding='utf-8')
```

```python
flights.isnull().sum()
```

```python
# Nao se tem info do aeroporto de destino e nao tem como inferir um
# Como sao so 2 registros, vamos dropar
flights = flights[flights['Aeroporto Destino'].notnull()]
flights.isnull().sum()
```

```python
# Há voos de um mesmo aeroporto para ele mesmo [?] (ex: 762108) com distancia percorrida 0
flights[flights['Passageiros Pagos'].isnull()].sort_values(['Ano', 'Mes'], ascending=False).head()
```

```python
flights_without_passanger = flights['Passageiros Pagos'].isnull().groupby([flights['Empresa'], flights['Ano'], flights['Mes']]).sum().astype(int).reset_index(name='count')
flights_without_passanger[flights_without_passanger['count'] > 0].nlargest(20, 'count')
```

```python
flights[flights['Assentos'].isnull()].sort_values(['Ano', 'Mes'], ascending=False).head()
```

```python
flights_without_passanger = flights['Assentos'].isnull().groupby([flights['Empresa'], flights['Ano'], flights['Mes']]).sum().astype(int).reset_index(name='count')
flights_without_passanger[flights_without_passanger['count'] > 0].nlargest(20, 'count')
```

```python
# Esses são os problemas encontrados, para solucionar usaremos a media dos outros da mesma empresa
# no mesmo trecho dos outros meses, caso não haja, usar de outra companhia

# Inicialmente, tirar estatísticas básicas por empresa/origem/destino/natureza/grupo de voo
# depois completar os valores com os correspondentes

# Mas primeiro vamos verificar nossa hipótese
```

```python
# Filling nulls with median
column_list = ['Passageiros Pagos', 'Passageiros Gratis', 'Carga Paga', 'Carga Gratis', 'Correio', 'Distancia', 'Decolagens', 'Assentos', 'Payload']
for i in column_list:
    flights[i] = flights.groupby(['Empresa', 'Aeroporto Origem', 'Aeroporto Destino', 'Natureza', 'Grupo de Voo'])[i].apply(lambda x: x.fillna(x.mean()))
```

```python
column_list = ['Passageiros Pagos', 'Passageiros Gratis', 'Carga Paga', 'Carga Gratis', 'Correio', 'Distancia', 'Decolagens', 'Assentos', 'Payload']
for i in column_list:
    flights[i] = flights.groupby(['Aeroporto Origem', 'Aeroporto Destino', 'Natureza', 'Grupo de Voo'])[i].apply(lambda x: x.fillna(x.mean()))
```

```python
column_list = ['Passageiros Pagos', 'Passageiros Gratis', 'Carga Paga', 'Carga Gratis', 'Correio', 'Distancia', 'Decolagens', 'Assentos', 'Payload']
for i in column_list:
    flights[i] = flights.groupby(['Natureza', 'Grupo de Voo'])[i].apply(lambda x: x.fillna(x.mean()))
```

```python
flights = flights.drop(flights.index[flights['Aeroporto Origem Pais'].isnull()].tolist())
flights = flights.drop(flights.index[flights['Aeroporto Destino Pais'].isnull()].tolist())
```

```python
flights.to_csv('flights_prepared.csv', header=1, index=False, encoding='utf-8')
```
